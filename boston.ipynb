```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
sns.set_style("whitegrid")
bostondata=load_boston()
df=pd.DataFrame(bostondata.data,columns=bostondata.feature_names)
from sklearn.preprocessing import StandardScaler
X=df[['LSTAT']].values
Y=bostondata.target
scx=StandardScaler()
scy=StandardScaler()
x=scx.fit_transform(X)
y=scy.fit_transform(Y.reshape(-1,1)).flatten()
n=100


w = np.zeros(1 + x.shape[1])
cost_ = []
alpha=0.0001
for i in range(n):
    y_pred = np.dot(x, w[1:]) + w[0]   #theta values are all stored in the w list 
    errors = (y - y_pred)
    w[1:] -= alpha * x.T.dot(errors)
    w[0] -= alpha * errors.sum()
    
    cost = (errors**2).sum() / 2.0
    cost_.append(cost)

plt.figure(figsize=(10,8))
plt.plot(range(1, n + 1), cost_);
plt.ylabel('SSE')
plt.xlabel('Epoch')
   
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_boston
from sklearn.tree import DecisionTreeRegressor

bostondata=load_boston()
df=pd.DataFrame(bostondata.data,columns=bostondata.feature_names)
sns.set_style("whitegrid")
df.head()

y=bostondata.target
df = pd.read_csv('housing.data', delim_whitespace=True, header=None)
df.columns = ['CRIM', 'ZN' , 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 
              'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
X = df[['LSTAT']].values
tree=DecisionTreeRegressor(max_depth=2)
tree.fit(X,y)
sort=X.flatten().argsort()
Y=tree.predict(X[sort])
plt.figure(figsize=(16,10))
plt.scatter(X[sort],y[sort])
plt.plot(X[sort],Y,color='k')
plt.xlabel('LSTAT')
plt.ylabel('MEDV')

```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor

x=df.values
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=42)
forest = RandomForestRegressor(n_estimators=500, criterion='mse', 
                               random_state=42, n_jobs=-1)
forest.fit(xtrain,ytrain)
ytrainpred = forest.predict(xtrain)
ytestpred = forest.predict(xtest)
print("MSE train: {0:.4f}, test: {1:.4f}".\
      format(mean_squared_error(ytrain, ytrainpred), 
             mean_squared_error(ytest, ytestpred)))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
from sklearn.linear_model import Ridge
from sklearn.model_selection import learning_curve
from sklearn.model_selection import ShuffleSplit

def plot_learning_curve(estimator, title, X, y, ylim=None, cv=None,
                        n_jobs=1, train_sizes=np.linspace(.1, 1.0, 5)):
    plt.figure(figsize=(10, 8))
    plt.xlabel("Training examples")
    plt.ylabel("Score")
    
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Cross-validation score")

    plt.legend(loc="best")
    return plt
boston = load_boston()
X, y = boston.data, boston.target

title = "Learning Curves (Ridge Regression)"

# Cross validation with 100 iterations to get smoother mean test and train
# score curves, each time with 20% data randomly selected as a validation set.

cv = ShuffleSplit(n_splits=100, test_size=0.2, random_state=0)
estimator = Ridge()
plot_learning_curve(estimator, title, X, y, cv=cv, n_jobs=4)
plt.show()
```
